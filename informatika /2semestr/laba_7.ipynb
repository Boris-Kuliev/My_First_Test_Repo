```python
import plotly
import plotly.graph_objs as go

import numpy as np

# Генерируем набор точек
x = np.arange(-2,2,0.01)
y = np.exp(x)+np.cos(x)

# Строим простейший график
fig1 = go.Figure()

# Создание объектов линий
gr1 = go.Scatter(x = x, y = y, name = 'График функции')

# Добавление линий на график
fig1.add_trace(gr1)

# Добавление курсора
fig1.update_traces(hoverinfo = 'x+y')

# Добавление маркеров (максимум)
ym = [y.max()]
xm = x[np.where(y == ym[0])]
mrk1 = go.Scatter(x = xm, y = ym, name = 'Max',
                  mode = 'markers+text',
                  marker = dict(size = 20, color = '#eb4044'),
                  text = 'fmax ('+str(xm[0])+','+str(ym[0])+')',
                  textposition = 'top center')
fig1.add_trace(mrk1)

# Подписи осей и графика
fig1.update_layout(title = 'График функции: y(t) = e**t+cos(t)',
                   xaxis_title = 't, с',
                   yaxis_title = 'f(t)',
                   legend_orientation = 'h')
fig1.update_traces(hoverinfo='all', hovertemplate='Аргумент: %{x}<br>Функция: %{y}')


fig1.show()
```

```python
import plotly
import plotly.graph_objs as go
import numpy as np

# Тестирование библиотеки plotly (карта Scattermapbox)

# Создаём объект "карта мира"
map1 = go.Scattermapbox(mode = 'lines+markers+text')

# Добавляем объект с картой на график
fig1 = go.Figure(map1)

# Рисуем линию на карте
fig1.add_trace(go.Scattermapbox(
    name = "трасса ВЛ",
    mode = 'lines',
    hoverinfo = 'skip',
    lat = [57.201, 57.26862, 57.26613, 57.12956, 57.12852, 57.14027, 57.11652, 57.12283, 57.1496, 57.128908, 57.130455, 57.130645, 57.132357, 57.126827, 57.109785, 57.1093876, 57.11],
    lon = [65.625, 65.36687, 65.28827, 64.64978, 63.81033, 63.66694, 63.54711, 62.26868, 62.12693, 61.769418, 61.764344, 61.760129, 61.687699, 61.683083, 61.6920021, 61.6939039, 61.71]))

# Добавление маркеров ПС
fig1.add_trace(go.Scattermapbox(
    name = "ПС Тюмень",
    mode = 'markers+text',
    marker = dict(size = 20, color = 'red'),
    text = ['ПС Тюмень 500 кВ'],
    lat = [57.201],
    lon = [65.625]))

fig1.add_trace(go.Scattermapbox(
    name = "ПС Рефтинская ГРЭС",
    mode = 'markers+text',
    marker = dict(size = 20, color = 'red'),
    text = ['ПС Рефтинская ГРЭС 500 кВ'],
    textposition = 'top center',
    lat = [57.11],
    lon = [61.71]))


# Добавляем стартовую точку обзора и способ вывода карты
mcent1 = go.layout.mapbox.Center(lat = 57.156, lon = 63.668)
fig1.update_layout(mapbox_style='open-street-map',
                   mapbox = dict(accesstoken = 'token.from.mapbox',
                                 center = mcent1,
                                 zoom = 8),
                   legend_orientation = 'h')

# Рисуем график (карту)
fig1.show()
```
